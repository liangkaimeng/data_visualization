# 柱状图

## 单数据系列柱状图

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")  # 设置画布背景
plt.rcParams['font.sans-serif'] = ['STKaiti']  # 设置中文不乱码
plt.rcParams['axes.unicode_minus'] = False  # 设置负号不乱码

fig = plt.figure(figsize=(8, 6), facecolor='none', edgecolor='none')  # 设置图形大小

# 下载数据
data = sns.load_dataset("titanic")

# 绘制柱状图
ax = sns.countplot(x=data["class"], orient='v', palette='Set2', alpha=1, edgecolor='green')

# 设置X轴刻度标签和标题
ax.set_xticklabels(['头等舱', '二等舱', '三等舱'], fontsize=8, color='black')
ax.set_xlabel("船舱等级", fontsize=10, color='black')

# 添加数值标签
ax.bar_label(ax.containers[0], label_type='edge', padding=-1, fontsize=11, rotation=None, color='red')

# 设置图形边缘空白最小化
fig.patch.set_alpha(0.0)

# 保存和显示图形
plt.savefig('../figures/seaborn_final_1.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 单数据系列柱状-折线图

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")  # 设置画布
plt.rcParams['font.sans-serif'] = ['STKaiti']  # 设置中文不乱码
plt.rcParams['axes.unicode_minus'] = False  # 设置负号不乱码

fig = plt.figure(figsize=(10, 5), facecolor='none', edgecolor='none')  # 设置图形大小

# 准备数据
data = sns.load_dataset("titanic")
group_df = (data.groupby("class").size() / float(len(data)) * 100).reset_index()
group_df[0] = group_df[0].apply(lambda x: round(x, 2))
group_df.columns = ['class', 'count']

# 绘制柱状图
ax1 = sns.countplot(x=data["class"], palette='Set2', alpha=1, edgecolor='green')

# 设置X轴刻度标签和标题
ax1.set_xticklabels(['头等舱', '二等舱', '三等舱'], fontsize=10)
ax1.set_xlabel("船舱等级", fontsize=10)

# 添加数值标签
ax1.bar_label(ax1.containers[0], label_type='edge', padding=-1, fontsize=11, rotation=None, color='red')

# 绘制折线图
ax2 = ax1.twinx()
lp = sns.lineplot(data=group_df, x="class", y='count', ax=ax2, marker='o', color='k', linewidth=0.7, markersize=3.5)

# 设置折线图的Y轴标签，隐藏折线图的Y轴刻度线
ax2.set_ylabel("占比 (%)", fontsize=10, color='k')
ax2.tick_params(axis='y', colors='k')

# 添加数值标签到折线图
for line in lp.lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        ax2.annotate(f'{int(y)}%', (x, y), ha='right', va='center', xytext=(0, -7), fontsize=8,
                     textcoords='offset points', color='k')

# 设置图形边缘空白最小化
fig.patch.set_alpha(0.0)

# 保存和显示图形
plt.savefig('../figures/seaborn_final_2.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 热力图

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")                   # 设置画布
plt.rcParams['font.sans-serif'] = ['STKaiti']  # 设置中文不乱码
plt.rcParams['axes.unicode_minus'] = False     # 设置负号不乱码

# 设置图形大小
fig = plt.figure(figsize=(8, 4), facecolor='none', edgecolor='none')

# 准备数据
glue = sns.load_dataset("glue").pivot("Model", "Task", "Score")

# 绘制图形
ax = sns.heatmap(glue, annot=True, fmt=".1f", linewidth=.1, cmap="twilight", annot_kws={"fontsize": 8})

# 修改横纵坐标字体大小
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)

# 设置图形边缘空白最小化
fig.patch.set_alpha(0.0)

# 保存和显示图形
plt.savefig('../figures/seaborn_final_28.png', dpi=500, bbox_inches='tight')
plt.show()
```
