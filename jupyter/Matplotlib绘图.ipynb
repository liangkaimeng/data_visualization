## 多变量柱状图

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['STSong']
plt.rcParams['axes.unicode_minus'] = False

fig = plt.figure(figsize=(8, 4), facecolor='none', edgecolor='none')  # 设置图形大小

# 数据
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([10, 15, 7, 12, 8])
y2 = np.array([5, 12, 8, 10, 6])
y3 = np.array([8, 9, 11, 6, 13])

# 柱状图的宽度
bar_width = 0.25

# 设置偏移量（x值）
x1 = x - bar_width
x2 = x
x3 = x + bar_width

# 绘制柱状图
plt.bar(x1, y1, width=bar_width, edgecolor='blue', label='Y1', color='cyan')
plt.bar(x2, y2, width=bar_width, edgecolor='blue', label='Y2', color='orange')
plt.bar(x3, y3, width=bar_width, edgecolor='blue', label='Y3', color='lime')

# 设置标签
plt.xlabel("X")
plt.ylabel("Y")

# 在每个柱子的顶部显示数值
for i, (x_val, y_val1, y_val2, y_val3) in enumerate(zip(x, y1, y2, y3)):
    plt.text(x_val - bar_width, y_val1 + 0.2, str(y_val1), ha='center', fontsize=9)
    plt.text(x_val, y_val2 + 0.2, str(y_val2), ha='center', fontsize=9)
    plt.text(x_val + bar_width, y_val3 + 0.2, str(y_val3), ha='center', fontsize=9)

# 设置图例
legend = plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3, fontsize=8)

# 设置图形边缘空白最小化
fig.patch.set_alpha(0.0)

# 保存和显示图形
plt.savefig('../figures/plt_final_5.png', dpi=200, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['STSong']  # 设置防止中文乱码
plt.rcParams['axes.unicode_minus'] = False  # 设置防止负号乱码

fig = plt.figure(figsize=(8, 4), facecolor='none', edgecolor='none')  # 设置图形大小

# 数据
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([10, 15, 7, 12, 8])
y2 = np.array([5, 12, 8, 10, 6])
y3 = np.array([8, 9, 11, 6, 13])

# 绘制堆叠柱状图
plt.bar(x, y1, label='Y1', edgecolor='blue', color='cyan')
plt.bar(x, y2, bottom=y1, label='Y2', edgecolor='blue', color='orange')
plt.bar(x, y3, bottom=y1+y2, label='Y3', edgecolor='blue', color='lime')

# 设置坐标轴标题与添加图例
plt.xlabel("X", fontsize=10)
plt.ylabel("Y", fontsize=10)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3, fontsize=8)

# 在每个柱子的内部显示数值
for i, (x_val, y_val1, y_val2, y_val3) in enumerate(zip(x, y1, y2, y3)):
    plt.text(x_val, y_val1/2, str(y_val1), ha='center', color='black')
    plt.text(x_val, y_val1 + y_val2/2, str(y_val2), ha='center', color='black')
    plt.text(x_val, y_val1 + y_val2 + y_val3/2, str(y_val3), ha='center', color='black')

# 设置图形边缘空白最小化
fig.patch.set_alpha(0.0)

# 保存和显示图形
plt.savefig('../figures/plt_final_8.png', dpi=200, bbox_inches='tight')
plt.show()
```

# 帕累托图

```python
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['STSong']  # 设置中文不乱码
plt.rcParams['axes.unicode_minus'] = False  # 设置负号不乱码

# 设置图形大小
fig = plt.figure(figsize=(8, 4), facecolor='none', edgecolor='none')

# 数据
categories = ['A', 'B', 'C', 'D', 'E']
counts = [30, 25, 15, 10, 5]

# 计算累计百分比
total = sum(counts)
cumulative_percent = [sum(counts[:i+1])/total*100 for i in range(len(counts))]

# 绘制帕累托图
fig, ax1 = plt.subplots()

# 绘制柱状图
bars = ax1.bar(categories, counts, color='skyblue')
ax1.set_ylabel('Counts')
ax1.set_ylim(0, 35)

# 添加数值标签
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, height, height, ha='center', va='bottom', fontsize=10)

# 绘制累计百分比折线图
ax2 = ax1.twinx()
ax2.plot(categories, cumulative_percent, color='lime', marker='^', markersize=4)
ax2.set_ylim(0, 110)
ax2.set_ylabel('Cumulative Percent')

# 添加折线图上的数值标签
for i, val in enumerate(cumulative_percent):
    ax2.text(categories[i], val, str(round(val, 2)) + '%', ha='center', va='bottom', fontsize=8)

# 设置标题和X轴标题
plt.title('Pareto Chart', fontsize=12)
plt.xlabel('Categories', fontsize=12)

# 设置图形边缘空白最小化
fig.patch.set_alpha(0.0)

# 保存和显示图形
plt.savefig('../figures/plt_final_13.png', dpi=200, bbox_inches='tight')
plt.show()
```

# 对比条形图

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['STSong']  # 设置中文不乱码
plt.rcParams['axes.unicode_minus'] = False  # 设置负号不乱码

# 设置图形大小
fig = plt.figure(figsize=(8, 4), facecolor='none', edgecolor='none')

# 创建子图
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), sharey=True)

# 创造数据
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [10, 15, 7, 12, 8]
values2 = [8, 12, 6, 9, 10]

# 绘制第一个子图
bar1 = ax1.barh(categories, values1, label='正例', edgecolor='blue', color='lime')
# 绘制第二个子图
bar2 = ax2.barh(categories, values2, label='负例', edgecolor='blue', color='cyan')

# 添加数值标签
for b in bar1:
    width = b.get_width()
    ax1.annotate(width, xy=(width, b.get_y() + b.get_height() / 2), xytext=(3, 0), 
                 textcoords='offset points', ha='left', va='center', fontsize=10)
    
for b in bar2:
    width = b.get_width()
    ax2.annotate(width, xy=(width, b.get_y() + b.get_height() / 2), xytext=(-3, 0), 
                 textcoords='offset points', ha='right', va='center', fontsize=10)

# 设置右边子图的刻度痕迹不可见
ax2.tick_params(axis='y', which='both', length=0, labelleft=False, labelright=False)

# 共享Y轴标签
fig.text(0.5, 0., 'Number', ha='center')

# 调整子图之间的间距
plt.subplots_adjust(wspace=0.)

# 反转X轴刻度
ax1.invert_xaxis()

# 设置图形边缘空白最小化
fig.patch.set_alpha(0.0)

# 保存和显示图形
plt.savefig('../figures/plt_final_16.png', dpi=200, bbox_inches='tight')
plt.show()
```

# 雷达图

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['STSong']  # 设置中文不乱码
plt.rcParams['axes.unicode_minus'] = False  # 设置负号不乱码

# 设置图形大小
fig = plt.figure(figsize=(8, 4), facecolor='none', edgecolor='none')

# 雷达图
def plot_radar(labels, data, score):
    n = len(labels)
    # 转化为十分制！！！
    if score in [5, 10, 100]:
        data = data * 10 / score
    elif score == 1:
        data = data * 10  
    angles = np.linspace(0, 2*np.pi, n, endpoint=False) 
    data = np.concatenate((data, [data[0]])) # 闭合
    angles = np.concatenate((angles, [angles[0]])) # 闭合
    
    fig = plt.figure()
    ax = fig.add_subplot(111, polar=True)# 参数polar，表示极坐标！！
    # 画grid线（5条环形线）
    for i in [2,4,6,8,10]:
        ax.plot(angles, [i]*(n+1), 'k-',lw=0.5,) # 之所以 n +1，是因为要闭合！
    
     # 填充底色
    ax.fill(angles, [10]*(n+1), facecolor='lightgreen', alpha=0.2)
    # 自己画grid线（n条半径线）
    for i in range(n):
        ax.plot([angles[i], angles[i]], [0, 10], 'b-',lw=0.5)
        
    # 画线(数据)!
    ax.set_theta_zero_location( "N")#从正上方开始(对称比较美@_@）
    ax.plot(angles, data, 'b-', linewidth=2, marker = "o", markersize =5)
    # 填充
    ax.fill(angles, data, facecolor='r', alpha=1.0)

    grid_angles = [ i* 360.0/n for i in range(n)]
    ax.set_thetagrids(grid_angles, labels, fontproperties="STSong")
    
    ax.set_rlim(0,10)
    # 下两行去掉所有默认的grid线
    ax.spines['polar'].set_visible(False) # 去掉最外围的黑圈
    ax.grid(False)                        # 去掉中间的黑圈

    ax.set_yticks([])# 关闭径向数值刻度
    
    # 设置图形边缘空白最小化
    fig.patch.set_alpha(0.0)
    
    file = '../figures/plt_final_17.png'
    plt.savefig(file, dpi=200, bbox_inches='tight')
    plt.show()
    
if __name__ == '__main__':
    labels = np.array(['守约', '人脉', '资产', '身份', '行为']) # 标签
    data = np.array([11, 60, 80, 60, 70]) # 数据
    score = 100 # 其可选的选项有1分制、5分制、10分制、100分制
    # 画雷达图
    plot_radar(labels, data, score)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['STSong']
plt.rcParams['axes.unicode_minus'] = False

# 设置图形大小
fig = plt.figure(figsize=(12, 4), facecolor='none', edgecolor='none')

# 创建数据
x = np.arange(1, 6)
y1 = np.array([10, 15, 7, 12, 8])
y2 = np.array([20, 18, 22, 17, 25])
y3 = np.array([5, 9, 6, 8, 10])

# 创建子图
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# 绘制柱状图
bars = axes[0].bar(x, y1, color='skyblue')
axes[0].set_title('Bar Chart')

# 在每个柱上显示数值
for bar in bars:
    height = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width() / 2, height,
                 str(height), ha='center', va='bottom')

# 绘制折线图
lines = axes[1].plot(x, y2, color='lime', marker='o')
axes[1].set_title('Line Chart')

# 绘制散点图
axes[2].scatter(x, y3, color='orange', marker='s')
axes[2].set_title('Scatter Chart')

# 调整子图之间的间距
plt.tight_layout()

# 设置图形边缘空白最小化
fig.patch.set_alpha(0.0)

# 保存图形
filename = '../figures/plt_final_20.png'
plt.savefig(filename, dpi=200, bbox_inches='tight')

# 显示图形
plt.show()
```
